費氏數列的遞迴表示式如下圖所示，但遞迴的程式在執行上缺乏效率，


請在下方的程式儲存格中設計一迭代 (iterative)程式以計算費氏數。

```python
# [20分]
# 請完成迭代的費氏數計算函式，並回傳答案。 輸入 n = 0,1,2,...
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
```

```python
# 請執行這一程式儲存格，但請不要做任何變更。
class Node():
    def __init__(self, data_=0):
        self.data = data_  # 儲存資料
        self.next = None   # 下一個節點參考

# Head 為一鏈結串列的head標籤(指標)
Head = Node(66)
```

請設計一函式對鏈結串列進行插入節點的動作

且插入的規則須滿足節點資料由小至大的排列順序

```python
# [20分]
# 請完成此自訂函式 insert()。 輸入H為鏈結串列的head，data為要插入節點的資料。
# 需回傳插入串列後的 head。
def insert(H, data):
    new_node = Node(data)
    if H is None or H.data > data:
        new_node.next = H
        return new_node
    Head = H
    while H.next != None and H.next.data < data:
        H = H.next
    new_node.next = H.next
    H.next = new_node
    return Head
```

```python
# 測試區間，請執行結果。
def traversal(H):
    ptr = H
    while(ptr!=None):
        print(ptr.data, end=' ')
        ptr = ptr.next

Head = insert(Head, 52)
Head = insert(Head, 36)
Head = insert(Head, 62)
Head = insert(Head, 88)
Head = insert(Head, 92)

traversal(Head)
```

__二元樹極其節點定義如下__

以二元樹的array表示法為輸入，以建構其對應的鏈結串列表示法

```python
# 請執行這一程式儲存格，但請不要做任何變更。
class Tree_Node():
    def __init__(self, data=0):
        self.Data = data
        self.Lchild = None
        self.Rchild = None

class Tree():
    def __init__(self, tree_list=[]):
        if(len(tree_list)>0):
            self.root = Tree.make_a_tree(1, tree_list) # idx從1開始
        else:
            self.root = None
    
    @staticmethod
    def make_a_tree(idx,L):
        if(idx<=len(L) and L[idx-1]!=None):
            new_node = Tree_Node(L[idx-1])
            new_node.Lchild = Tree.make_a_tree(2*idx, L)
            new_node.Rchild = Tree.make_a_tree(2*idx+1, L)
            return new_node
        
T1 = Tree(['A', 'B', 'C', None, 'D', 'E', 'I', None, None, 'F', 'G', 'H'])
T2 = Tree(['A', 'B', 'F', 'C', None, 'G', None, None, 'D', None, None, 'E'])
T2 = Tree(['A', 'B', 'F', 'C', None, 'G', None, None, 'D', None, None, None, None, None, None, None, None, 'E'])
```

請設計一函式對對此樹計算節點個數

```python
# [20分]
# 請完成此自訂函式 node_count()。 
# 輸入ptr為計算該樹節點個數的樹根(T.root)。函式回傳對應樹的節點數。
def node_count(ptr):
    if ptr != None:
        NL = node_count(ptr.Lchild)
        NR = node_count(ptr.Rchild)
        return NL + NR + 1
    else:
        return 0
```

```python
# 測試區間，請執行結果。
print(f'T1有{node_count(T1.root)}個節點')
print(f'T3有{node_count(T2.root)}個節點')
```

請設計一函式對對此樹的數高 (__定義: 樹根的樹高為1__)

```python
# [20分]
# 請完成此自訂函式 tree_height()。 
# 輸入ptr為計算該樹樹高的樹根(T.root)。函式回傳對應樹的樹高。
def tree_height(ptr):
    if ptr == None:
        return 0
    NL = tree_height(ptr.Lchild)
    NR = tree_height(ptr.Rchild)
    return max(NL , NR) + 1 
```

```python
# 測試區間，請執行結果。
print(f'T1的高度為{tree_height(T1.root)}')
print(f'T3的高度為{tree_height(T2.root)}')
```
